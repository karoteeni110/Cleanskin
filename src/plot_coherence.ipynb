```python

import pandas as pd
import matplotlib.pyplot as plt

def toCSV(fpath):
    with open(fpath,'r') as f:
        stats = f.readlines()

    print("#topics,coherence")
    for i in range(0,42,3):
        topics = stats[i].strip()
        coh = stats[i+1][-8:-2]
        print(topics+","+coh)

toCSV('./130k_coh')
```

```python
# Print data

df = pd.read_csv('./6k_coh',dtype={"#topics":int, "coherence":float})
# print(df)
df.sort_values(by="#topics",inplace=True)
# df.drop([8,9,10,11],inplace=True)
print(df)
```

```python
# Plot coherence

ax = df.plot.scatter(x="#topics",y="coherence") #,xticks=df["#topics"])
plt.annotate('max: (70, 0.6459)', # 'max: (30, 0.6482)', 
        # xy=(30, 0.6482), xycoords='data',
        # xytext=(40, 0.60), textcoords='data',
        xy=(70, 0.6459), xycoords='data',
        xytext=(90, 0.60), textcoords='data',
        size=10, va="center",
        bbox=dict(boxstyle="round", fc=(1.0, 0.7, 0.7), ec="none"),
        arrowprops=dict(arrowstyle="wedge,tail_width=1.",
                        fc=(1.0, 0.7, 0.7), ec="none",
                        relpos=(0.2, 0.5)))
plt.show()
```
